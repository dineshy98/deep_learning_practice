```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense,Activation
```

```python
df = pd.read_csv(
    "https://data.heatonresearch.com/data/t81-558/iris.csv")
df.head()
```

```python
X = df.iloc[:,0:4].values
y = pd.get_dummies(df['species'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test  = train_test_split(X,y)
```

```python
X.shape
```

```python
#Building Neural Network
model = Sequential()
model.add(Dense(50,input_dim = X.shape[1],activation = 'relu'))
model.add(Dense(25,activation = 'relu'))
model.add(Dense(y.shape[1],activation = 'softmax'))
model.compile(loss = 'categorical_crossentropy',metrics=['accuracy'],optimizer = 'adam')
```

```python
history = model.fit(X_train,y_train,epochs = 200)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.plot(history.history['accuracy'])
plt.show()
```

```python
#Building Neural Network
model1 = Sequential()
model1.add(Dense(50,input_dim = X.shape[1],activation = 'relu'))
model1.add(Dense(25,activation = 'relu'))
model1.add(Dense(y.shape[1],activation = 'softmax'))
model1.compile(loss = 'categorical_crossentropy',metrics=['accuracy'],optimizer = 'adam')
```

```python
history1 = model1.fit(X_train,y_train,validation_data = (X_test,y_test),epochs = 1000)
```

```python
k = pd.DataFrame(history1.history)
```

```python
plt.plot(k.loss)
plt.plot(k.val_loss)
plt.legend(['Train', 'Test'])
```

```python
plt.plot(k.accuracy)
plt.plot(k.val_accuracy)
plt.legend(['Train', 'Test'])
```

```python
#Building Neural Network
model2 = Sequential()
model2.add(Dense(50,input_dim = X.shape[1],activation = 'relu'))
model2.add(Dense(25,activation = 'relu'))
model2.add(Dense(y.shape[1],activation = 'softmax'))
model2.compile(loss = 'categorical_crossentropy',metrics=['accuracy'],optimizer = 'adam')
```

```python
from keras.callbacks import EarlyStopping
monitor = EarlyStopping(monitor = 'val_loss',min_delta = 1e-3,patience = 5,
                        verbose= 1,mode = 'auto',restore_best_weights = True)
history2 = model2.fit(X_train,y_train,validation_data = (X_test,y_test),epochs = 1000,callbacks = [monitor])
```

```python
l = pd.DataFrame(history2.history)
plt.plot(l.loss)
plt.plot(l.val_loss)
plt.legend(['Train', 'Test'])
```

```python
plt.plot(l.accuracy)
plt.plot(l.val_accuracy)
plt.legend(['Train', 'Test'])
```
